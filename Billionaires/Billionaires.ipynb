
# Lets analys the data 
#### data column header is too big. If we reduce size we may see clearly.

### lets shorform long headers with short length column
* company.founded = CF
* company.relationship = CR
* company.sector = CS
* company.type = CT
* demographics.age = DA
* demographics.gender = DG
* location.citizenship = LC
* location.country code = LCC
* wealth.worth in billions =WWIB
* wealth.how.category = WHC
* wealth.how.from emerging = WHFE
* wealth.how.inherited = WHI
* wealth.how.was founder = WHWF
* wealth.how.was political = WHWP

```python
raw_data.rename(columns = {'company.founded':'CF', 'company.relationship':'CR', 'company.type':'CT', 'demographics.age':'DA', 'demographics.gender':'DG', 'company.sector':'CS',
                           'wealth.worth in billions':'WWIB', 'wealth.how.category':'WHC', 'wealth.how.from emerging':'WHFE', 'wealth.how.inherited':'WHI', 'location.citizenship':'LC',
                           'wealth.how.was founder':'WHWF', 'wealth.how.was political':'WHWP', 'location.country code':'LCC' }, inplace = True)
```

```python
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)

print(raw_data.head(5))
```

### Lets remove unwanted column

```python
halfcleaned_data= raw_data.drop(['LCC', 'WHFE', 'location.gdp', 'location.region', 'WHWF', 'WHWP', 'WHC', 'CT', 'WHI', 'wealth.how.industry'], axis=1)
print(halfcleaned_data.head(20))
```

### lets remove row with with dupplicate name

```python
print(len(halfcleaned_data[halfcleaned_data.duplicated('name')]))
```

```python
filter_data = halfcleaned_data.drop_duplicates(['name'], keep='last')
```

```python
print(len(filter_data))
```

```python
print(filter_data)
```

```python
filter_data['Rank'] = range(1, 2077 + 1 ,1)
filter_data
```

```python
clean_data = filter_data.drop(['rank'], axis=1)
```

```python
clean_data.set_index(['Rank'], inplace = True)
clean_data
```
